```python
!pip install beautifulsoup4 --upgrade --quiet
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(page_contents,'html.parser')
```

```python
soup
```

```python
selection_class = 'f3 lh-condensed mb-0 mt-1 Link--primary'
topic_title_tags = soup.find_all('p', {'class':selection_class})
```

```python
len(topic_title_tags)
```

```python
topic_title_tags[:5]
```

```python
desc_selector = 'f5 color-fg-muted mb-0 mt-1'
topic_desc_tags = soup.find_all('p', {'class':desc_selector})
```

```python
topic_desc_tags[:5]
```

```python
topic_title_tag0 = topic_title_tags[0]

topic_title_tag0
```

```python
# to see whats inside the tags (.text)
topic_title_tags[0].text
```

 Now we can find the two things --

```python
# print out the topic and its desc along with the parents tags

topic_title_tag0.parent.parent
```

#### We wants get   link Address for the 3D

```python
topic_link_tags =soup.find_all('a', {'class':'no-underline flex-1 d-flex flex-column'})
```

```python
len(topic_link_tags)
```

```python
topic_link_tags[:1]
```

```python
topic_link_tags[0]['href']
```

```python
topic0_url ="https://github.com/" + topic_link_tags[0]['href']
print(topic0_url)
```

#### Now we Enter the 3D page Sections

```python
# Here is the All list of topics

topic_titles=[]

for tag in topic_title_tags:
    topic_titles.append(tag.text)

print(topic_titles)
```

```python
# strip() is used for removing the extra space 
topic_desc = []

for tag in topic_desc_tags:
    topic_desc.append(tag.text.strip())
    
topic_desc[:5]
```

```python
topic_urls = []
base_url = 'https://github.com'

for tag in topic_link_tags:
    topic_urls.append(base_url + tag['href'])
    
topic_urls
```

```python
topic_dict = {
    
    'title':topic_titles,
    'description': topic_desc,
    'urls': topic_urls
}
```
